```python
import numpy as py
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("nothing.csv")
```

```python
df.head()
```

```python
plt.scatter(df.age, df.life, marker='*', color = 'red')
plt.show
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[['age']], df.life, train_size=0.9, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_some = model.predict(x_test)
score = accuracy_score(y_test , y_some)
score
```
